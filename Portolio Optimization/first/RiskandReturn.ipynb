# Risk and Return of a Porfolio

## Theoretical Concepts

A portfolio is a diversified collection of financial assets, encompassing stocks, bonds, commodities, and various investment instruments, held by individuals or entities. Its primary purpose is to spread risk across different assets, allowing investors to achieve a balance between potential returns and risk. The construction of portfolios is driven by investors' financial goals, risk tolerance, and investment strategy.

Harry Markowitz, awarded the Nobel Prize in Economics in 1990 for his influential work on Mean-Variance Analysis, laid the foundation for Modern Portfolio Theory (MPT). This groundbreaking theory transformed the approach to portfolio construction by underscoring the importance of considering both return and risk.

Mean-Variance Analysis defines portfolio return as the expected capital gain and portfolio risk as potential capital losses due to adverse market movements. The analysis utilizes the mean and variance of the portfolio's fluctuations to model these factors comprehensively. In essence, Modern Portfolio Theory integrates these principles, guiding investors in constructing well-balanced portfolios that align with their financial objectives and risk preferences.

### Portfolio Return:

The portfolio return $r$ is determined by the weighted summation of individual asset returns:

$$
r = \sum_{i=1}^{n} w_i \cdot r_i 
$$

where $n$ is the number of assets with weights $w_i$ and $r_i$ as the asset returns.

Then, the annualized portfolio return, assuming $T$ trading days in a year, is given by:

$$ 
R = T \sum_{i=1}^{n} w_i \cdot r_i 
$$

### Portfolio Risk:

Portfolio risk is the standard deviation of its returns:

$$
\sigma_P = \sqrt{\sum_{i=1}^{n} \sum_{j=1}^{n} w_i \cdot w_j \cdot \sigma_{ij}} 
$$

where $\sigma_{ij}$ is the covariance of returns between assets $i$ and $j$ of the portfolio, also known as the variance-covariance matrix of returns.

The annualized portfolio risk in percentage, $\sigma_{P\%}$, is given by:

$$ 
\sigma_{P\%} = \sqrt{T} \cdot \sigma_P 
$$

where $T$ is the number of trading days in a year.

### Risk-Return Tradeoff

The Risk-Return Tradeoff is a core principle in finance, illustrating the connection between potential returns and associated risks. It involves the inherent dilemma that pursuing higher returns typically involves taking on greater risks, and vice versa. Investors must navigate this tradeoff to construct portfolios aligned with their financial goals and risk tolerance. Balancing the desire for higher returns with an acceptable level of risk is essential for informed investment decision-making.
